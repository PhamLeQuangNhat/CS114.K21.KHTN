```
from imutils import paths
import os
import numpy as np 
import cv2
from preprocessing.simplepreprocessor import SimplePreprocessor
from dataloader.simpledatasetloader import SimpleDatasetLoader
import h5py
```

```
imagePaths = list(paths.list_images("/content/drive/My Drive/ML/CS114.K21.KHTN/Final/Data/Train_data(1)"))
```

```
print(len(imagePaths))
```

```
sp = SimplePreprocessor(32,32)
sdl = SimpleDatasetLoader(preprocessors=[sp])
(data, labels) = sdl.load(imagePaths, verbose=100)
data = data.astype("float") / 255.0
data = data.reshape((data.shape[0], 3072))
```

```
print(data.shape)
```

```
print(labels.shape)
```

```
h5_data          = 'H5PY/data.h5'
h5_labels        = 'H5PY/labels.h5'
```

```
h5f_data = h5py.File(h5_data, 'w')
h5f_data.create_dataset('dataset', data=np.array(data))
```

```
from sklearn.preprocessing import LabelEncoder
```

```
label = LabelEncoder().fit_transform(labels)
```

```
print(label.shape)
```

```
print(label[0])
```

```
h5f_label = h5py.File(h5_labels, 'w')
h5f_label.create_dataset('dataset', data=np.array(label))
```

```
h5f_data.close()
h5f_label.close()
```

```
from sklearn.svm import SVC
```

```
data  = h5py.File(h5_data, 'r')
label = h5py.File(h5_labels, 'r')
```

```
dataset = data['dataset']
```

```
dataset = np.array(dataset)
```

```
Labels = label['dataset']
```

```
Labels = np.array(Labels)
```

```
print(dataset.shape)
```

```
print(Labels.shape)
```

```
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
```

```
# partition the data into training: 75%, testing: 25%
(trainX, testX, trainY, testY) = train_test_split(dataset, Labels, test_size=0.25, random_state=42)
```

```
model = SVC()
model.fit(trainX, trainY)
```

```
# evaluate a SVM 
print("[INFO] evaluating SVM...")
predictions = model.predict(testX)

```

```
print(predictions[0])
```

```
print(classification_report(testY,predictions))
```
