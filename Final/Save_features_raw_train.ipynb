# **4. Save features**

## **4.1 Import các thư viện cần thiết**

```
from dataloader.simpledatasetloader import SimpleDatasetLoader
from preprocessing.simplepreprocessor import SimplePreprocessor
from Feature_Extraction.hog import HOG
from Feature_Extraction.localbinarypatterns import LocalBinaryPatterns

import numpy as np 
import h5py
import os
from imutils import paths
from sklearn.preprocessing import LabelEncoder
```

## **4.2 Save features train**

```
imagePaths = 'Data/Raw_train_data'
```

```
# grab the list of images 
print("[INFO] loading images...")
imagePaths_train = list(paths.list_images(imagePaths)) 
```

```
print(len(imagePaths_train))
```



### **4.2.1 Normal vector**

```
data_raw_normal = 'H5PY/raw_train/train_normal.h5'
labels_raw_h5    = 'H5PY/raw_train/labels_train.h5'
```

```
# initialize image preprocessor
sp = SimplePreprocessor(32,32)
sdl = SimpleDatasetLoader(preprocessors=[sp])

# load dataset from memory
(data, labels) = sdl.load(imagePaths_train,verbose=100)
```

```
# scale the raw pixel intensities to the range [0, 1]
data = data.astype("float") / 255.0
```

```
# reshape the data matrix
data = data.reshape((data.shape[0],3072))
```

```
# encode the labels as integers
le = LabelEncoder()
labels = le.fit_transform(labels)
```

```
# save the feature vector using HDF5
h5f_data = h5py.File(data_raw_normal, 'w')
h5f_data.create_dataset('dataset', data=np.array(data))
```

```
h5f_label = h5py.File(labels_raw_h5, 'w')
h5f_label.create_dataset('dataset', data=np.array(labels))
```

```
data.shape
```

```
labels.shape
```

```
h5f_data.close()
h5f_label.close()
```

###**4.2.2 Local Binary Patterns**

```
data_raw_lbps = 'H5PY/raw_train/train_lbps.h5'
```

```
# initialize image preprocessor
sp = SimplePreprocessor(32,32)

# initialize local binary patterns
lbps = LocalBinaryPatterns(24,8)

sdl = SimpleDatasetLoader(preprocessors=[sp,lbps])

# load dataset from memory
(data, labels) = sdl.load(imagePaths_train,verbose=100)
```

```
data = data.astype("float")
```

```
data.shape
```

```
# save the feature vector using HDF5
h5f_data = h5py.File(data_raw_lbps, 'w')
h5f_data.create_dataset('dataset', data=np.array(data))
```

```
h5f_data.close()
```
